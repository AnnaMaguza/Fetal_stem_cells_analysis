#### **Datasets**

In this notebiook we will prepare the following datasets for the annotation process:

- **E-MTAB-8901** (Elementaite et al., 2021) - Fetal developing gut
- **E-MTAB-9536** (Elementaite et al., 2021) - Fetal developing gut
- **E-MTAB-9543** (Elementaite et al., 2021) - Adult gut
- **E-MTAB-9489** (Holloway et al., 2021) - Fetal developing gut
- **E-MTAB-9720** (Holloway et al., 2021) - Fetal enteroids

Here we will conduct such preprocessing:

1. **Cell Annotation Transfer**

   - Cells present in both the Gut Cell Atlas (Elementaite, 2021) and the following Elementaite datasets (E-MTAB-8901, E-MTAB-9536, and E-MTAB-9543) are assigned cell type labels consistent with the Gut Cell Atlas annotations.
   - Cells from the Holloway datasets (E-MTAB-9489 and E-MTAB-9720) and unannotated cells from the Elementaite datasets are initially labeled as 'Unknown'.

2. **Datasets merge**

3. **Visualization of uncorrected PCA and UMAP plots to identify main covariates**

## Prepare obs

+ Make barcodes as in the GCA dataset

```python
E8901.obs.rename(columns={E8901.obs.columns[0]: 'cell_id'}, inplace=True)
E9489.obs.rename(columns={E9489.obs.columns[0]: 'cell_id'}, inplace=True)
E9536.obs.rename(columns={E9536.obs.columns[1]: 'cell_id'}, inplace=True)
E9543.obs.rename(columns={E9543.obs.columns[1]: 'cell_id'}, inplace=True)
E9720.obs.rename(columns={E9720.obs.columns[0]: 'cell_id'}, inplace=True)
```

```python
anndata_objects = [E8901, E9489, E9536, E9543, E9720]
```

```python
for adata in anndata_objects:
    adata.obs['cell_id'] = adata.obs['cell_id'].str.split('-').str[0]
```

```python
anndata_objects = [E8901, E9536, E9543]
```

```python
for adata in anndata_objects:
    adata.obs['batch'] = adata.obs['batch'].astype(str)
    adata.obs['cell_id'] = adata.obs['cell_id'] + '-1-' + adata.obs['batch']
    adata.obs['barcode'] = adata.obs.index.copy()
    adata.obs['barcode'] = adata.obs['barcode'].str.split('-').str[0]
    adata.obs.index = adata.obs['cell_id']
```

```python
common_indexes = E8901.obs.index.intersection(GCA.obs.index)
print(f"Number of common indexes: {len(common_indexes)}")
common_indexes = E9536.obs.index.intersection(GCA.obs.index)
print(f"Number of common indexes: {len(common_indexes)}")
common_indexes = E9543.obs.index.intersection(GCA.obs.index)
print(f"Number of common indexes: {len(common_indexes)}")
```

```python
anndata_objects = [E9489, E9720]

for adata in anndata_objects:
    adata.obs['Source Name'] = adata.obs['Source Name'].astype(str)
    adata.obs['cell_id'] = adata.obs['cell_id'] + '-1-' + adata.obs['Source Name']
    adata.obs['barcode'] = adata.obs.index.copy()
    adata.obs['barcode'] = adata.obs['barcode'].str.split('-').str[0]
    adata.obs.index = adata.obs['cell_id']
```

+ Merge cell types and cell states from GCA to Elementaite objects, cells that are not present in GCA have annotation 'Unknown'

```python
anndata_objects = [E8901, E9536, E9543]
```

```python
for adata in anndata_objects:
    adata.obs['category'] = 'Unknown'
    adata.obs['Integrated_05'] = 'Unknown'
    
    common_indexes = adata.obs.index.intersection(GCA.obs.index)
    
    adata.obs.loc[common_indexes, 'category'] = GCA.obs.loc[common_indexes, 'category']
    adata.obs.loc[common_indexes, 'Integrated_05'] = GCA.obs.loc[common_indexes, 'Integrated_05']
```

```python
anndata_objects = [E9489, E9720]

for adata in anndata_objects:
    adata.obs['category'] = 'Unknown'
    adata.obs['Integrated_05'] = 'Unknown'
```

```python
del GCA # to reduce memory usage
```

## Merge datasets

```python
anndata_objects = {'E-MTAB-9720': E9720, 'E-MTAB-E8901': E8901, 'E-MTAB-E9489': E9489, 'E-MTAB-E9543': E9543, 'E-MTAB-E9536': E9536}
```

```python
combined_processing_history = []

for key, adata in anndata_objects.items():
    if 'processing_history' in adata.uns:
        for entry in adata.uns['processing_history']:
            combined_processing_history.append(f'[{key}] {entry}')
```

```python
merged_data = anndata.concat(
    list(anndata_objects.values()), 
    join='outer', 
    label=None, 
    keys=list(anndata_objects.keys()), 
    index_unique=None, 
    uns_merge='unique',
)
```

```python
merged_data.uns['processing_history'] = np.array(combined_processing_history)
print(merged_data.uns['processing_history'])
```

+ Filter out genes that do not expressed

```python
sc.pp.filter_genes(merged_data, min_counts=1)
merged_data
```

```python
del anndata_objects # to reduce memory usage
```

## Fix observations

```python
merged_data.obs['Protocol REF'] = merged_data.obs['Protocol REF'].cat.add_categories('Unknown')
merged_data.obs['Protocol REF'] = merged_data.obs['Protocol REF'].replace(['  '], 'Unknown')
```

```python
merged_data.obs['library_construction'] = merged_data.obs['library_construction'].astype(str)
merged_data.obs['LIBRARY_LAYOUT'] = merged_data.obs['LIBRARY_LAYOUT'].astype(str)
merged_data.obs['library_construnction_and_layout'] = merged_data.obs['library_construction'] + '_' + merged_data.obs['LIBRARY_LAYOUT']
merged_data.obs['library_construnction_and_layout'].value_counts()
```

```python
merged_data.obs['donor_id'] = merged_data.obs['donor_id'].cat.add_categories('cell_culture')
merged_data.obs['donor_id'] = merged_data.obs['donor_id'].replace(['nan'], 'cell_culture')
```

+ Fix issue with Boolean None

```python
nan_summary = merged_data.obs.isna().sum()

print(nan_summary[nan_summary > 0])
```

```python
nan_info = nan_summary[nan_summary > 0].to_frame(name='NaN Count')
nan_info['Data Type'] = merged_data.obs.dtypes[nan_summary > 0]

print(nan_info)
```

```python
merged_data.obs['cell_type'] = merged_data.obs['cell_type'].replace(['  '], 'unsorted')
merged_data.obs['cell_type'] = merged_data.obs['cell_type'].replace(['epithelial cell'], 'intestinal epithelial cell')
merged_data.obs['cell_type'] = merged_data.obs['cell_type'].cat.remove_unused_categories()
merged_data.obs['cell_type'] = merged_data.obs['cell_type'].fillna('unsorted')
```

```python
merged_data.obs['cluster_passed_qc'] = merged_data.obs['cluster_passed_qc'].astype(str)
merged_data.obs['cluster_passed_qc'] = merged_data.obs['cluster_passed_qc'].fillna('N/A')
merged_data.obs['consensus_fraction'] = merged_data.obs['consensus_fraction'].astype(str)
merged_data.obs['consensus_fraction'] = merged_data.obs['consensus_fraction'].fillna('N/A')
merged_data.obs['consensus_passed_qc'] = merged_data.obs['consensus_passed_qc'].astype(str)
merged_data.obs['consensus_passed_qc'] = merged_data.obs['consensus_passed_qc'].fillna('N/A')
merged_data.obs['qc_cluster'] = merged_data.obs['qc_cluster'].astype(str)
merged_data.obs['qc_cluster'] = merged_data.obs['qc_cluster'].fillna('N/A')
```

```python
merged_data.obs['gestational_age'] = merged_data.obs['gestational_age'].cat.add_categories('N/A - not fetal')
merged_data.obs['gestational_age'] = merged_data.obs['gestational_age'].replace(['  '], 'N/A - not fetal')
merged_data.obs['gestational_age'] = merged_data.obs['gestational_age'].fillna('N/A - not fetal')
```

```python
merged_data.obs['passage'] = merged_data.obs['passage'].cat.add_categories('N/A')
merged_data.obs['passage'] = merged_data.obs['passage'].replace(['  '], 'N/A')
merged_data.obs['passage'] = merged_data.obs['passage'].fillna('N/A')
```

```python
merged_data.obs['batch'] = merged_data.obs['batch'].cat.add_categories('unknown')
merged_data.obs['batch'] = merged_data.obs['batch'].fillna('unknown')
```

```python
merged_data.obs['sampling_site'] = merged_data.obs['sampling_site'].cat.add_categories('unknown')
merged_data.obs['sampling_site'] = merged_data.obs['sampling_site'].replace(['  '], 'unknown')
merged_data.obs['sampling_site'] = merged_data.obs['sampling_site'].fillna('unknown')
```
