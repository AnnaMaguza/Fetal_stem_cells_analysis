This notebook is created to analyze the batch correction and bioconservation of integrated healthy reference gut dataset using `scIB` metrics.

```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata as an
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scib_metrics.benchmark import Benchmarker
```

```python
adata = sc.read_h5ad('Healthy_reference/Integrated/Integrated_4_datasets_05042024.h5ad')
```

+ Study name as batch

```python
bm = Benchmarker(
    adata,
    batch_key='Study_name',
    label_key='Cell_Type',
    embedding_obsm_keys=['X_scANVI', 'X_scVI', 'X_pca'],
    n_jobs=-1
)
bm.benchmark()
```

```python
bm.plot_results_table(min_max_scale=False)
```
