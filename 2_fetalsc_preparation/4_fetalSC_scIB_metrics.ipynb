This notebook is created to analyze the batch correction of fetal stem cells with `scIB` metrics

```python
import numpy as np
import pandas as pd
import scanpy as sc
import anndata as an
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from scib_metrics.benchmark import Benchmarker
```

```python
adata = sc.read_h5ad('data/Fetal_stem_cells_leiden.h5ad')
```

```python
adata
```

```python
bm = Benchmarker(
    adata,
    batch_key='Donor_ID',
    label_key='cluster',
    embedding_obsm_keys=['X_scVI', 'X_pca'],
    n_jobs=-1
)
bm.benchmark()
```

```python
bm.plot_results_table(min_max_scale=False)
```
